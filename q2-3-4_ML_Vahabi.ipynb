### 3-a-normalized

```python
import statistics
def normalize(in_vector):
    rv = []
    total_mean = statistics.mean(in_vector)
    total_std = np.std(in_vector)
    for i in range(len(in_vector)):
        rv.append(in_vector[i] - total_mean / total_std)
    return rv
```

### 3-b-discretize

```python
def find_bin(vl, s, total_min, num_Bins):
    total_bin=0
    for i in range(num_Bins):
        if vl <= total_min+(i*s):
            total_bin = i
            break
    return total_bin
```

```python
def discretize_attribute(in_vector, bins):
    in_max = max(in_vector)
    in_min = min(in_vector)
    bin_size = (in_max - in_min) / bins
    output = []
    for i in range(len(in_vector)):
        tv = find_bin(in_vector[i], bin_size, in_min, bins)
        output .append(tv)
    return output
```

```python
input_vector = pima_data['BloodPressure']
input_vector=input_vector.to_numpy()
discretize_attribute(input_vector,20)
```

```python
import random
def divideset1(ds, p_train):
    training = []
    testing = []
    for i in range(len(ds)):
        r=np.random.random((1, 1))
        if r < p_train:
            training.append(ds.iloc[i])
        else:
            testing.append(ds.iloc[i])
    return (training,testing)
```

```python
num_iteration=20
p_train=0.66
len_training=[]
len_testing=[]
for i in range(num_iteration):
    out=divideset1(pima_data, p_train)
    #print(len(out[0])+len(out[1]))
    len_training.append(len(out[0]))
    len_testing.append(len(out[1]))
```

```python
print('training size list:') 
print(len_training) 
print('testing size list:') 
print(len_testing) 
sumtrlen=sum(len_training)
avgtrlen=sumtrlen/num_iteration
print('\n')
print('training avg list:'+str(avgtrlen)) 
```

### 4_c

```python
import numpy as np
def divideset2(ds, p_train):
    training = []
    testing = []
    dataset_size = len(ds)
    training_size = dataset_size * p_train; 
    trn_counter = 1
    used_data = np.zeros((dataset_size, 1))

    while trn_counter < training_size:
        d=[i for i in range(dataset_size)]
        rnd = random.choice(d)
        if used_data[rnd] == 0:
            used_data[rnd] = 1
            trn_counter = trn_counter + 1
            training.append(ds.iloc[rnd])
        
    for i in range(dataset_size):
        if used_data[i] == 0:
            testing.append(ds.iloc[i])
            
    return (training,testing)
       
```

```python
num_iteration=20
p_train=0.66
len_training=[]
len_testing=[]
for i in range(num_iteration):
    out=divideset2(pima_data, p_train)
    #print(len(out[0])+len(out[1]))
    len_training.append(len(out[0]))
    len_testing.append(len(out[1]))
```

```python
print('training size list:' + str(len_training[0])) 
print('testing size list:' + str(len_testing[0])) 
print('training list:') 
print(out[0]) 
print('testing list:') 
print(out[1]) 
```

### 4-Validation scheme -a


```python
def kfold_crossvalidation(ds, k, m):
    mode_one = np.mod(len(ds), k)
    floor_grp = np.floor(len(ds)/k)
    sizes = np.zeros(k)
    
    for i in range(k):
        if mode_one > 0 :
            sizes[i] = floor_grp + 1
            mode_one = mode_one - 1
        else:
            sizes[i] = floor_grp
        
    training = []
    testing = []
    start_index = 1
    for j in range(k):
        if j == m:
            
            
            x=start_index + sizes[j] - 1
            x= int(x)
            start_index= int(start_index)
            #print(ds[start_index : x])

            testing.append(ds[start_index : x])
        else:
            #print(type(sizes(j)))
            x=start_index + sizes[j] - 1
            x= int(x)
            start_index= int(start_index)
            #print(ds[start_index : x])

            training.append(ds[start_index : x])
        start_index = start_index + sizes[j]
    return(training,testing)
    
```

```python
out_kfold=kfold_crossvalidation(pima_data, 20, 10)
print('Training set is:')
print(out_kfold[0])
print('\n\n\n')
print('Testing set is:')
print(out_kfold[1])
```

### 4-Validation scheme - b


```python
def bootstrap1(ds, k):
    random_sample_ind = random.sample(range(0, len(ds)), k)
    sample_data = []
    for i in range(k):
        sample_data.append(ds.iloc[random_sample_ind[i]])
    return sample_data
    
```
